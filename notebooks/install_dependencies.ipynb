## 7. 驗證安裝


```python
# 檢查已安裝的套件
required_packages = [
    'pandas',
    'numpy',
    'requests',
    'yfinance',
    'tqdm',
    'scikit-learn',
    'statsmodels',
    'scipy',
    'matplotlib',
    'seaborn'
]

print("檢查已安裝的套件：")
print("=" * 50)

for package in required_packages:
    try:
        __import__(package.replace('-', '_'))
        print(f"✓ {package}")
    except ImportError:
        print(f"✗ {package} (未安裝)")

# 檢查 TA-Lib（可能安裝失敗）
try:
    import talib
    print(f"✓ TA-Lib")
except ImportError:
    print(f"✗ TA-Lib (未安裝 - 可能需要手動安裝)")
    print("  提示：Windows 上安裝 TA-Lib 可能需要額外步驟")
    print("  參考：https://github.com/TA-Lib/ta-lib-python")
```
